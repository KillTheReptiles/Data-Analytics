```python
### paquetes de impotacion y manejo de datos
import pandas as pd
import numpy as np

 

# ### paquetes de graficos
import matplotlib.pyplot as plot
# from scipy import stats

 

 

# ######  paquetes de analitica de datos
# from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

 


archivo='abalone.csv'

 

pepito=pd.read_csv(archivo)
Nombre_columnas=['Sex', 
'Length',
'Diameter',
'Height',
'Whole weight' ,
'Shucked weight', 
'Viscera weight', 
'Shell weight ',
'Rings',]
pepito.columns=Nombre_columnas

plot.hist(x=pepito['Diameter'])
plot.subplots()
plot.hist(x=pepito['Height'])
plot.subplots()
plot.boxplot(x=pepito['Viscera weight'])
plot.subplots()
plot.scatter(x=pepito['Rings'],y=pepito['Height'])
#plot.hist(x=pepito['Sex'])
X=pepito['Height']
Y=pepito['Rings']

X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        Y,
                                        train_size   = 0.5,
                                    )
```

```python
modelo = LinearRegression()
modelo.fit(X = np.array(X_train).reshape(-1,1), y = y_train)
predicciones = modelo.predict(X = np.array(X_test).reshape(-1,1))
rmse = mean_squared_error(y_true = y_test, y_pred = predicciones)

print('El valor valor de RMSE es: ', rmse)

predicciones = modelo.predict(X = np.array(X_test).reshape(-1,1))
r2 = r2_score(y_true = y_test, y_pred = predicciones)
print('El valor valor de r2 es: ', r2)
```
